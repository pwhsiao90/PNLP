## Find new keyword by using N-Gram

```python
with open("data/corpus/travel_01.txt", "r", encoding='utf-8') as file:
    lines = file.readlines()
```

```python
lines[200:210]
```

```python
articles = [x.replace('\n', '').replace('\u3000\u3000','') for x in lines if (len(x.strip()) >1)]
```

```python
articles[200:210]
```

```python
from collections import Counter

SPECIAL_CHARS = '❤️→★☆▌!"#$%&()*+,-./:;<=＝>?@[\]^_`─{|}~！，。 「」？、～1234567890１２３４５６７８９０:《》：】【）『』（QAZWSXEDCRFVTGBYHNUJMIKOLPqazwsxedcrfvtgbyhnujmikolp'
MIN_OCCURANCE = 10

original_dict_path = 'dict/dict.big.txt'
original_word_list = []
file = open(original_dict_path, 'r', encoding='utf-8')
for line in file:
    original_word_list.append(line.split(' ')[0])
    
def ngrams(article, n):
    output = []
    for i in range(len(article)-n+1):
        CLEAN = True
        for c in article[i:i+n]:
            if c in SPECIAL_CHARS:
                CLEAN = False
                break
        if CLEAN:
            output.append(article[i:i+n])
    return output

def get_new_words_new(article, maxlength=5):
    maxlength += 1
    ngram = []
    for i in range(maxlength+1):
        ngram.append(ngrams(article, i+1))
    
    article_length = len(ngram[0])
    min_count = max(article_length*4/1000, 5)
    
    ngram_counter = []
    for i in range(maxlength+1):
        ngram_counter.append(Counter(ngram[i]))
    
    ngram_counter_f = []
    threshold_array_ratio = [1.3] + [1]*maxlength
    for i in range(maxlength+1):
        ngram_counter_f.append({x : ngram_counter[i][x] for x in ngram_counter[i] if ngram_counter[i][x] >= min_count*threshold_array_ratio[i]})
    
    new_words = []
    n_m_gram_ratio = [0.8, 0.9, 0.9] + [0.9]*(maxlength-2)
    for i in range(1, maxlength):
        # i+1 gram
        for word, count in ngram_counter_f[i].items():
            is_new = False
            for gram in ngrams(word, i):
                if ngram_counter_f[i][word] /ngram_counter[i-1][gram] >= n_m_gram_ratio[i-1] :
                    is_new = True
                    if gram in new_words:
                        new_words.remove(gram)
            
            if is_new and len(word) < maxlength and word not in original_word_list:
                new_words.append(word)
    
    return new_words
```

```python
for idx, article in enumerate(articles[200:210]):
    words = get_new_words_new(article,5)
    if len(words) > 0: print('{}:{}'.format(idx,words))
```

### Add to custom dictionary

```python
from tqdm import tqdm

w_ = []
new_words = []
for article in tqdm(articles[200:210]):
    words = get_new_words_new(article,5)
    if len(words) > 0:
        w_.append(words)

for x in tqdm(w_):
    for y in x:
        new_words.append(y)
```

```python
import pandas as pd

df_ = pd.DataFrame(new_words)
df_.head()
```

```python
# Write data into dict/mydic_new_words.xlsx
writer = pd.ExcelWriter('dict/mydic_new_words.xlsx', options={'strings_to_urls': False})
df_.to_excel(writer,'New words', index = True)
writer.save()
```
