## Create training set from ld+json

```python
def get_ld_json_from_url(uri):
    rs = requests.session()
    header = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/64.0.3282.167 Safari/537.36'}
    res = rs.get(uri, headers=header, timeout=300)
    html_data = res.text
    l = html_data.split('<script type="application/ld+json">')[1].split('</script>')[0]
    j = json.dumps(json.loads(l), ensure_ascii=False)
    jObj = json.loads(j)
    return jObj
```
